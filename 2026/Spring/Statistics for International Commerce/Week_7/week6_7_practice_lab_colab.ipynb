## Part 2. Confidence interval with sigma known

### Task 2A
Compute the standard error.

```python
xbar = 94
sigma = 21
n = 49
z_star = 1.96

se = sigma / np.sqrt(n)
se
```

### Task 2B
Compute the margin of error.

```python
me = z_star * se
me
```

## Part 4. Small-sample business interpretation

A duty-free retailer samples 25 tourists and finds:

- sample mean spending = 68 USD
- sample standard deviation = 15 USD

Construct a 90% confidence interval for mean tourist spending.

```python
xbar = 68
s = 15
n = 25
df = n - 1
t_star = t.ppf(0.95, df=df)  # 90% CI

se = s / np.sqrt(n)
me = t_star * se
lower = xbar - me
upper = xbar + me

pd.DataFrame({
    "SE": [se],
    "t*": [t_star],
    "ME": [me],
    "Lower": [lower],
    "Upper": [upper]
})
```

## Part 6. Mini case: global e-commerce decision

A global e-commerce company is considering whether to advertise “average delivery in under 5 days.”

Recent data:
- sample mean delivery time = 5.4 days
- sample standard deviation = 1.2 days
- sample size = 36

Use a 95% confidence interval to assess this claim.

```python
xbar = 5.4
s = 1.2
n = 36
df = n - 1
t_star = t.ppf(0.975, df=df)

se = s / np.sqrt(n)
me = t_star * se
lower = xbar - me
upper = xbar + me

pd.DataFrame({
    "SE": [se],
    "t*": [t_star],
    "ME": [me],
    "Lower": [lower],
    "Upper": [upper]
})
```
