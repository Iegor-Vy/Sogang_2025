## Part 1. z-test: average export order value

A company claims that the average export order value is 100 USD.

A sample shows:
- x-bar = 108
- population sigma = 20
- n = 25

Use a two-sided test at alpha = 0.05.

### Task 1A
State the hypotheses.

### Task 1B
Compute the standard error.

```python
xbar = 108
mu0 = 100
sigma = 20
n = 25

se = sigma / np.sqrt(n)
se
```

### Task 1C
Compute the z statistic.

```python
z_stat = (xbar - mu0) / se
z_stat
```

### Task 1D
Compute the two-sided p-value.

```python
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
p_value
```

## Part 3. t-test: average shipping surcharge

A company believes the average express shipping surcharge is 40 USD.

A sample of 16 shipments gives:
- x-bar = 42
- s = 8
- n = 16

Use a two-sided t-test at alpha = 0.05.

### Task 3A
Why do we use t instead of z here?

### Task 3B
Compute the standard error.

```python
xbar = 42
mu0 = 40
s = 8
n = 16

se = s / np.sqrt(n)
se
```

### Task 3C
Compute the t statistic.

```python
t_stat = (xbar - mu0) / se
t_stat
```

### Task 3D
Compute the two-sided p-value.

```python
df = n - 1
p_value = 2 * (1 - t.cdf(abs(t_stat), df))
p_value
```

### Your answer

## Part 4. Confidence interval and hypothesis test

A company samples order processing times and gets:
- x-bar = 60
- sigma = 12
- n = 36

At the 5% significance level, management wants to test:

H0: mu = 65
H1: mu != 65

Instead of testing directly, use a 95% confidence interval.

### Task 4A
Construct the 95% confidence interval.

```python
xbar = 60
sigma = 12
n = 36
z_star = 1.96

se = sigma / np.sqrt(n)
me = z_star * se
lower = xbar - me
upper = xbar + me

lower, upper
```

### Task 4B
Based on the interval, should H0: mu = 65 be rejected?
