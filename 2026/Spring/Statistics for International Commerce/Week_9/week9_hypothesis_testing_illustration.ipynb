## 1. Big idea

A business claim might say:

"Average delivery time is under 5 days."

We collect a sample and ask:
- Is the sample result consistent with the claim?
- Or is it too unusual under that claim?

That is the logic of hypothesis testing.

## 2. Null and alternative hypotheses

Example:

A logistics company claims the average delivery time is 5 days.

We write:
- H0: mu = 5
- H1: mu != 5

or
- H0: mu = 5
- H1: mu > 5

depending on the business question.

## 3. Visual intuition for a test statistic

Suppose H0 says the population mean is 100.

If our sample mean is far from 100, the test statistic becomes large in absolute value.

```python
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)

plt.plot(x, y, linewidth=2)
plt.axvline(0, linestyle="--", color="black", label="center under H0")
plt.axvline(2, linestyle="--", color="red", label="example test statistic")
plt.axvline(-2, linestyle="--", color="red")
plt.title("Test Statistic on the Standard Normal Curve")
plt.xlabel("z")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## 4. Example: z-test for mean order value

A company claims:
- average order value = 100 USD

Sample information:
- x-bar = 108
- sigma = 20
- n = 25

Because population sigma is known, we use a z-test.

```python
xbar = 108
mu0 = 100
sigma = 20
n = 25

se = sigma / np.sqrt(n)
z_stat = (xbar - mu0) / se
z_stat
```

```python
p_value_two_sided = 2 * (1 - norm.cdf(abs(z_stat)))
p_value_two_sided
```

Interpretation:
- z tells us how many standard errors the sample mean is from the null mean
- p-value tells us how unusual that is if H0 were true

## 5. Shade the p-value area

```python
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, linewidth=2)
plt.fill_between(x[x >= abs(z_stat)], y[x >= abs(z_stat)], alpha=0.3)
plt.fill_between(x[x <= -abs(z_stat)], y[x <= -abs(z_stat)], alpha=0.3)
plt.axvline(z_stat, linestyle="--", color="red")
plt.axvline(-z_stat, linestyle="--", color="red")
plt.title("Two-Sided p-value Area")
plt.xlabel("z")
plt.ylabel("Density")
plt.show()

print(f"z statistic = {z_stat:.3f}")
print(f"two-sided p-value = {p_value_two_sided:.4f}")
```

## 6. Example: t-test for shipping surcharge

A company wants to test whether the average express shipping surcharge is 40 USD.

Sample data:
- x-bar = 42
- s = 8
- n = 16

Because sigma is unknown, we use a t-test.

```python
xbar = 42
mu0 = 40
s = 8
n = 16
df = n - 1

se = s / np.sqrt(n)
t_stat = (xbar - mu0) / se
t_stat
```

```python
p_value_t = 2 * (1 - t.cdf(abs(t_stat), df))
p_value_t
```

## 7. Compare z and t distributions

```python
x = np.linspace(-4, 4, 500)

plt.plot(x, norm.pdf(x), label="z distribution", linewidth=2)
plt.plot(x, t.pdf(x, df=5), label="t distribution (df=5)", linewidth=2)
plt.plot(x, t.pdf(x, df=20), label="t distribution (df=20)", linewidth=2)
plt.title("z vs t")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

Main point:
- t is wider for small samples
- t approaches z as sample size grows

## 8. Decision rule using alpha

At significance level alpha = 0.05:
- if p-value < 0.05, reject H0
- if p-value >= 0.05, do not reject H0

```python
alpha = 0.05

print("z-test:")
print("Reject H0" if p_value_two_sided < alpha else "Do not reject H0")

print("\nt-test:")
print("Reject H0" if p_value_t < alpha else "Do not reject H0")
```

## 9. Type I and Type II error

Type I error:
- reject H0 when H0 is true

Type II error:
- do not reject H0 when H0 is false

Business examples:
- Type I: changing pricing policy when no true difference exists
- Type II: missing a real rise in delivery time

## 10. Confidence interval and hypothesis test connection

For a two-sided test at alpha = 0.05:
- if the null value lies outside the 95% confidence interval -> reject H0
- if the null value lies inside the 95% confidence interval -> do not reject H0

```python
xbar = 60
sigma = 12
n = 36
z_star = 1.96

se = sigma / np.sqrt(n)
me = z_star * se
lower = xbar - me
upper = xbar + me

lower, upper
```

If H0 says mu = 65, then 65 is outside the interval, so H0 would be rejected at the 5% level.
