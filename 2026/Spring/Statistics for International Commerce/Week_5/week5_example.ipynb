## Week 5 Interactive Visualizations

This notebook covers:
- Continuous probability distributions
- Uniform distribution
- Normal distribution
- Z-scores and probabilities
- Normal approximation to the binomial

Use the sliders to explore how the distributions change.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, binom
import ipywidgets as widgets
from ipywidgets import interact

plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['font.size'] = 12
```

## 1. Uniform Distribution

```python
def plot_uniform(a=0, b=30, shade_left=10, shade_right=20):
    if b <= a:
        print('b must be greater than a')
        return

    x = np.linspace(a - (b-a)*0.2, b + (b-a)*0.2, 1000)
    y = np.where((x >= a) & (x <= b), 1/(b-a), 0)

    plt.figure()
    plt.plot(x, y, linewidth=2)

    left = max(a, min(shade_left, shade_right))
    right = min(b, max(shade_left, shade_right))
    if left < right:
        xs = np.linspace(left, right, 400)
        ys = np.full_like(xs, 1/(b-a))
        plt.fill_between(xs, ys, alpha=0.3)
        prob = (right - left) / (b-a)
    else:
        prob = 0

    mean = (a+b)/2
    std = (b-a)/np.sqrt(12)

    plt.axvline(mean, linestyle='--', label=f'Mean = {mean:.2f}')
    plt.title('Uniform Distribution')
    plt.xlabel('X')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    print(f'Mean = {mean:.2f}')
    print(f'Standard deviation = {std:.2f}')
    print(f'P({min(shade_left,shade_right):.1f} < X < {max(shade_left,shade_right):.1f}) = {prob:.4f}')

interact(
    plot_uniform,
    a=widgets.FloatSlider(value=0, min=-10, max=20, step=1, description='a'),
    b=widgets.FloatSlider(value=30, min=1, max=50, step=1, description='b'),
    shade_left=widgets.FloatSlider(value=10, min=-10, max=50, step=1, description='left'),
    shade_right=widgets.FloatSlider(value=20, min=-10, max=50, step=1, description='right')
)
```

## 2. Normal Distribution

```python
def plot_normal(mu=1000, sigma=100, x1=840, x2=1200):
    if sigma <= 0:
        print('sigma must be positive')
        return

    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = norm.pdf(x, mu, sigma)

    plt.figure()
    plt.plot(x, y, linewidth=2)

    lo, hi = min(x1, x2), max(x1, x2)
    xs = np.linspace(lo, hi, 500)
    ys = norm.pdf(xs, mu, sigma)
    plt.fill_between(xs, ys, alpha=0.3)

    z1 = (x1 - mu)/sigma
    z2 = (x2 - mu)/sigma
    prob = norm.cdf(hi, mu, sigma) - norm.cdf(lo, mu, sigma)

    plt.axvline(mu, linestyle='--', label=f'Mean = {mu:.2f}')
    plt.title('Normal Distribution')
    plt.xlabel('X')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    print(f'z({x1:.1f}) = {z1:.2f}')
    print(f'z({x2:.1f}) = {z2:.2f}')
    print(f'P({lo:.1f} < X < {hi:.1f}) = {prob:.4f}')

interact(
    plot_normal,
    mu=widgets.FloatSlider(value=1000, min=0, max=2000, step=50, description='μ'),
    sigma=widgets.FloatSlider(value=100, min=10, max=400, step=10, description='σ'),
    x1=widgets.FloatSlider(value=840, min=0, max=2000, step=10, description='x1'),
    x2=widgets.FloatSlider(value=1200, min=0, max=2000, step=10, description='x2')
)
```

## 3. Z-Score Explorer

```python
def zscore_explorer(x=790, mu=1000, sigma=100):
    if sigma <= 0:
        print('sigma must be positive')
        return

    z = (x - mu)/sigma
    prob = norm.cdf(x, mu, sigma)

    xx = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    yy = norm.pdf(xx, mu, sigma)

    plt.figure()
    plt.plot(xx, yy, linewidth=2)
    xs = np.linspace(xx.min(), x, 500)
    ys = norm.pdf(xs, mu, sigma)
    plt.fill_between(xs, ys, alpha=0.3)
    plt.axvline(x, linestyle='--', label=f'x = {x:.1f}')
    plt.title('Z-score and Left-tail Probability')
    plt.xlabel('X')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    print(f'Z = {z:.3f}')
    print(f'P(X < {x:.1f}) = {prob:.4f}')

interact(
    zscore_explorer,
    x=widgets.FloatSlider(value=790, min=0, max=2000, step=10, description='x'),
    mu=widgets.FloatSlider(value=1000, min=0, max=2000, step=50, description='μ'),
    sigma=widgets.FloatSlider(value=100, min=10, max=400, step=10, description='σ')
)
```

## 4. Normal Approximation to the Binomial

```python
def binomial_vs_normal(n=80, p=0.7, k=60):
    if p <= 0 or p >= 1:
        print('p must be between 0 and 1')
        return

    x = np.arange(0, n+1)
    binom_pmf = binom.pmf(x, n, p)

    mu = n*p
    sigma = np.sqrt(n*p*(1-p))

    xx = np.linspace(0, n, 1000)
    normal_pdf = norm.pdf(xx, mu, sigma)

    plt.figure()
    plt.bar(x, binom_pmf, alpha=0.5, label='Binomial PMF')
    plt.plot(xx, normal_pdf, linewidth=2, label='Normal Approximation')
    plt.axvline(k, linestyle='--', label=f'k = {k}')
    plt.title('Binomial vs Normal Approximation')
    plt.xlabel('X')
    plt.ylabel('Probability / Density')
    plt.legend()
    plt.show()

    exact = 1 - binom.cdf(k-1, n, p)
    approx = 1 - norm.cdf(k-0.5, mu, sigma)

    print(f'Mean = np = {mu:.2f}')
    print(f'Std = sqrt(np(1-p)) = {sigma:.2f}')
    print(f'Condition check: np = {n*p:.2f}, n(1-p) = {n*(1-p):.2f}')
    print(f'Exact binomial P(X ≥ {k}) = {exact:.4f}')
    print(f'Normal approx with continuity correction = {approx:.4f}')

interact(
    binomial_vs_normal,
    n=widgets.IntSlider(value=80, min=10, max=200, step=5, description='n'),
    p=widgets.FloatSlider(value=0.7, min=0.05, max=0.95, step=0.05, description='p'),
    k=widgets.IntSlider(value=60, min=0, max=200, step=1, description='k')
)
```

## 5. Empirical Rule (68–95–99.7)

```python
def empirical_rule(mu=1000, sigma=100):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = norm.pdf(x, mu, sigma)

    plt.figure()
    plt.plot(x, y, linewidth=2)

    for k in [1, 2, 3]:
        xs = np.linspace(mu-k*sigma, mu+k*sigma, 400)
        ys = norm.pdf(xs, mu, sigma)
        plt.fill_between(xs, ys, alpha=0.15)

    plt.axvline(mu, linestyle='--', label='Mean')
    plt.title('Empirical Rule for Normal Distribution')
    plt.xlabel('X')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    print(f'Within 1σ: {norm.cdf(mu+sigma, mu, sigma)-norm.cdf(mu-sigma, mu, sigma):.4f}')
    print(f'Within 2σ: {norm.cdf(mu+2*sigma, mu, sigma)-norm.cdf(mu-2*sigma, mu, sigma):.4f}')
    print(f'Within 3σ: {norm.cdf(mu+3*sigma, mu, sigma)-norm.cdf(mu-3*sigma, mu, sigma):.4f}')

interact(
    empirical_rule,
    mu=widgets.FloatSlider(value=1000, min=0, max=2000, step=50, description='μ'),
    sigma=widgets.FloatSlider(value=100, min=10, max=400, step=10, description='σ')
)
```
