# Part 1. Do customers spend differently across regions?

An international e-commerce company wants to know whether average customer spending differs across three regions:
- Asia
- Europe
- North America

This is a suitable ANOVA question because we compare more than two group means.

```python
asia = np.array([72, 68, 75, 80, 66, 70, 73, 78, 69, 74, 71, 77])
europe = np.array([61, 65, 63, 67, 60, 66, 62, 64, 68, 59, 65, 63])
north_america = np.array([82, 78, 85, 88, 80, 84, 86, 79, 83, 87, 81, 85])

df_spending = pd.DataFrame({
    "region": ["Asia"] * len(asia) + ["Europe"] * len(europe) + ["North America"] * len(north_america),
    "spending_usd": np.concatenate([asia, europe, north_america])
})

df_spending.head()
```

## Task 1A. Descriptive statistics

Compute the mean and standard deviation of spending for each region.

```python
df_spending.groupby("region")["spending_usd"].agg(["count", "mean", "std"])
```

## Task 1B. Visualization

```python
groups = [
    df_spending[df_spending["region"] == "Asia"]["spending_usd"],
    df_spending[df_spending["region"] == "Europe"]["spending_usd"],
    df_spending[df_spending["region"] == "North America"]["spending_usd"]
]

plt.boxplot(groups, labels=["Asia", "Europe", "North America"])
plt.title("Customer Spending by Region")
plt.ylabel("Spending (USD)")
plt.show()
```

# Part 3. Do shipping costs differ by shipping mode?

A global retailer uses three shipping modes:
- Air
- Sea
- Land

The company wants to know whether the average shipping cost differs across these modes.

```python
air = np.array([28, 32, 35, 30, 33, 31, 36, 29, 34, 37])
sea = np.array([12, 15, 14, 13, 16, 11, 15, 14, 12, 13])
land = np.array([18, 20, 19, 22, 21, 17, 23, 20, 19, 21])

df_shipping = pd.DataFrame({
    "shipping_mode": ["Air"] * len(air) + ["Sea"] * len(sea) + ["Land"] * len(land),
    "shipping_cost_usd": np.concatenate([air, sea, land])
})

df_shipping
```

## Task 3A. Compute group summaries

```python
df_shipping.groupby("shipping_mode")["shipping_cost_usd"].agg(["count", "mean", "std"])
```

## Task 3B. Visualize the data

```python
groups = [
    df_shipping[df_shipping["shipping_mode"] == "Air"]["shipping_cost_usd"],
    df_shipping[df_shipping["shipping_mode"] == "Sea"]["shipping_cost_usd"],
    df_shipping[df_shipping["shipping_mode"] == "Land"]["shipping_cost_usd"]
]

plt.boxplot(groups, labels=["Air", "Sea", "Land"])
plt.title("Shipping Cost by Shipping Mode")
plt.ylabel("Shipping Cost (USD)")
plt.show()
```

## Task 3C. Run ANOVA

```python
f_stat, p_value = stats.f_oneway(air, sea, land)
f_stat, p_value
```

# Part 4. Important Limitation: ANOVA does not tell us which groups differ

ANOVA answers:

Is at least one group mean different?

It does not directly answer:

Which exact groups are different?

To answer that, we need follow-up comparisons.

## Task 4A. Pairwise comparisons

Run simple pairwise t-tests below.

Note: In more advanced courses, we would adjust for multiple testing. Here, we use this only for learning.

```python
pairs = [
    ("Air vs Sea", air, sea),
    ("Air vs Land", air, land),
    ("Sea vs Land", sea, land)
]

for name, group1, group2 in pairs:
    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)
    print(name)
    print("t-statistic:", round(t_stat, 3))
    print("p-value:", round(p_val, 4))
    print()
```

# Part 5. Do product ratings differ across product categories?

A company sells three product categories:
- Electronics
- Fashion
- Beauty

Management wants to know whether customer ratings differ across categories.

```python
electronics = np.array([4.2, 4.4, 4.1, 4.5, 4.3, 4.0, 4.6, 4.2, 4.4, 4.3])
fashion = np.array([3.8, 3.9, 4.0, 3.7, 4.1, 3.8, 3.9, 4.0, 3.6, 3.9])
beauty = np.array([4.5, 4.6, 4.4, 4.7, 4.5, 4.8, 4.6, 4.7, 4.4, 4.6])

df_ratings = pd.DataFrame({
    "category": ["Electronics"] * len(electronics) + ["Fashion"] * len(fashion) + ["Beauty"] * len(beauty),
    "rating": np.concatenate([electronics, fashion, beauty])
})

df_ratings.head()
```

## Task 5A. Group summaries

```python
df_ratings.groupby("category")["rating"].agg(["count", "mean", "std"])
```

## Task 5B. Run ANOVA

```python
f_stat, p_value = stats.f_oneway(electronics, fashion, beauty)
f_stat, p_value
```
