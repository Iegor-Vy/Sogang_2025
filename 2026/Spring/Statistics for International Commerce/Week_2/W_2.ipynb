## Assignment 3. Inflation and Household Purchasing Power

### Goal
Assess whether nominal wage growth is keeping up with inflation.

### Questions
1. Compute the mean nominal wage growth.
2. Compute the mean inflation rate.
3. Compute the mean real wage growth.
4. In which years was real wage growth negative?
5. Did household purchasing power improve or worsen on average?


```python
df_wages = pd.DataFrame({
    "Year": [2019, 2020, 2021, 2022, 2023],
    "Nominal_Wage_Growth_pct": [3.0, 2.5, 4.0, 5.0, 4.5],
    "Inflation_pct": [1.5, 2.0, 3.8, 6.5, 5.2]
})

df_wages["Real_Wage_Growth_pct"] = df_wages["Nominal_Wage_Growth_pct"] - df_wages["Inflation_pct"]

df_wages
```

```python
df_wages.describe()
```

```python
df_wages[df_wages["Real_Wage_Growth_pct"] < 0]
```
