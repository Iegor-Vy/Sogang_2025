## 1. Setup

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from scipy.stats import binom, hypergeom, poisson
import ipywidgets as widgets
from IPython.display import display
```

## 2. A simple probability distribution

We begin with a custom discrete probability distribution.
This helps students understand:
- outcomes
- probabilities
- expected value
- variance

```python
x_vals = np.array([0, 1, 2, 3, 4])
p_vals = np.array([0.10, 0.25, 0.30, 0.20, 0.15])

mu = np.sum(x_vals * p_vals)
var = np.sum((x_vals - mu)**2 * p_vals)
sd = np.sqrt(var)

df = pd.DataFrame({
    "x": x_vals,
    "P(x)": p_vals
})

df
```

```python
fig = go.Figure()
fig.add_bar(x=x_vals, y=p_vals)
fig.add_vline(x=mu, line_dash="dash", annotation_text=f"Mean = {mu:.2f}", annotation_position="top right")
fig.update_layout(
    title="Custom Probability Distribution",
    xaxis_title="Outcome",
    yaxis_title="Probability",
    template="plotly_white"
)
fig.show()

print(f"Expected value = {mu:.2f}")
print(f"Variance = {var:.3f}")
print(f"Standard deviation = {sd:.3f}")
```

## 3. Interactive binomial distribution

Economic interpretation:
Let X = number of delayed shipments in a week.

- n = number of shipments
- p = probability of delay

You can change n and p and see:
- how the shape changes
- how the expected number of delays changes
- how risk changes

```python
def plot_binomial(n=10, p=0.2):
    x = np.arange(0, n + 1)
    y = binom.pmf(x, n, p)
    mu = n * p
    var = n * p * (1 - p)

    fig = go.Figure()
    fig.add_bar(x=x, y=y, name="P(X=x)")
    fig.add_vline(x=mu, line_dash="dash", annotation_text=f"Mean = {mu:.2f}", annotation_position="top right")
    fig.update_layout(
        title=f"Binomial Distribution: n={n}, p={p}",
        xaxis_title="Number of successes",
        yaxis_title="Probability",
        template="plotly_white"
    )
    fig.show()

    print(f"Expected value = {mu:.2f}")
    print(f"Variance = {var:.3f}")
    print(f"P(X=0) = {binom.pmf(0, n, p):.4f}")
    print(f"P(X>=1) = {1 - binom.pmf(0, n, p):.4f}")
    print(f"P(X<=2) = {binom.cdf(2, n, p):.4f}")

widgets.interact(
    plot_binomial,
    n=widgets.IntSlider(value=10, min=1, max=30, step=1, description="n"),
    p=widgets.FloatSlider(value=0.2, min=0.01, max=0.95, step=0.01, description="p")
)
```

## 4. Interactive hypergeometric distribution

Economic interpretation:
Let X = number of defective items found in a sample.

- N = total number of items in the batch
- K = number of defective items in the batch
- n = number sampled without replacement

This model is useful for:
- quality control
- audit sampling
- committee selection

```python
def plot_hypergeom(N=30, K=8, n=5):
    if K > N or n > N:
        print("Please choose values with K <= N and n <= N.")
        return

    max_x = min(K, n)
    x = np.arange(0, max_x + 1)
    y = hypergeom.pmf(x, N, K, n)
    mu = n * (K / N)

    fig = go.Figure()
    fig.add_bar(x=x, y=y, name="P(X=x)")
    fig.add_vline(x=mu, line_dash="dash", annotation_text=f"Mean = {mu:.2f}", annotation_position="top right")
    fig.update_layout(
        title=f"Hypergeometric Distribution: N={N}, K={K}, n={n}",
        xaxis_title="Number of successes in sample",
        yaxis_title="Probability",
        template="plotly_white"
    )
    fig.show()

    print(f"Expected value = {mu:.2f}")
    print(f"P(X=0) = {hypergeom.pmf(0, N, K, n):.4f}")
    print(f"P(X>=1) = {1 - hypergeom.pmf(0, N, K, n):.4f}")

widgets.interact(
    plot_hypergeom,
    N=widgets.IntSlider(value=30, min=10, max=100, step=1, description="N"),
    K=widgets.IntSlider(value=8, min=1, max=30, step=1, description="K"),
    n=widgets.IntSlider(value=5, min=1, max=20, step=1, description="n")
)
```

## 5. Interactive Poisson distribution

Economic interpretation:
Let X = number of rare events in a fixed interval.

Examples:
- lost bags per flight
- document errors per day
- customer complaints per hour

The key parameter is lambda, the average rate.

```python
def plot_poisson(lam=1.5, xmax=12):
    x = np.arange(0, xmax + 1)
    y = poisson.pmf(x, lam)

    fig = go.Figure()
    fig.add_bar(x=x, y=y, name="P(X=x)")
    fig.add_vline(x=lam, line_dash="dash", annotation_text=f"Mean = {lam:.2f}", annotation_position="top right")
    fig.update_layout(
        title=f"Poisson Distribution: lambda={lam}",
        xaxis_title="Number of events",
        yaxis_title="Probability",
        template="plotly_white"
    )
    fig.show()

    print(f"Expected value = {lam:.2f}")
    print(f"Variance = {lam:.2f}")
    print(f"P(X=0) = {poisson.pmf(0, lam):.4f}")
    print(f"P(X<=1) = {poisson.cdf(1, lam):.4f}")

widgets.interact(
    plot_poisson,
    lam=widgets.FloatSlider(value=1.5, min=0.1, max=10.0, step=0.1, description="lambda"),
    xmax=widgets.IntSlider(value=12, min=3, max=30, step=1, description="x max")
)
```

## 6. Compare distributions side by side

This section helps students see the difference between the three distributions.

Suggested interpretation:
- Binomial: repeated independent trials with fixed p
- Hypergeometric: sampling without replacement
- Poisson: rare counts in time or space

```python
def compare_distributions(n=10, p=0.2, N=30, K=8, n_sample=5, lam=2.0):
    if K > N or n_sample > N:
        print("Please choose values with K <= N and n_sample <= N.")
        return

    x_bin = np.arange(0, n + 1)
    y_bin = binom.pmf(x_bin, n, p)

    max_h = min(K, n_sample)
    x_h = np.arange(0, max_h + 1)
    y_h = hypergeom.pmf(x_h, N, K, n_sample)

    x_p = np.arange(0, max(12, int(lam * 4)) + 1)
    y_p = poisson.pmf(x_p, lam)

    fig = go.Figure()
    fig.add_scatter(x=x_bin, y=y_bin, mode="lines+markers", name="Binomial")
    fig.add_scatter(x=x_h, y=y_h, mode="lines+markers", name="Hypergeometric")
    fig.add_scatter(x=x_p, y=y_p, mode="lines+markers", name="Poisson")
    fig.update_layout(
        title="Comparison of Discrete Distributions",
        xaxis_title="x",
        yaxis_title="Probability",
        template="plotly_white"
    )
    fig.show()

widgets.interact(
    compare_distributions,
    n=widgets.IntSlider(value=10, min=3, max=25, step=1, description="binom n"),
    p=widgets.FloatSlider(value=0.2, min=0.01, max=0.95, step=0.01, description="binom p"),
    N=widgets.IntSlider(value=30, min=10, max=80, step=1, description="hyper N"),
    K=widgets.IntSlider(value=8, min=1, max=20, step=1, description="hyper K"),
    n_sample=widgets.IntSlider(value=5, min=1, max=15, step=1, description="hyper n"),
    lam=widgets.FloatSlider(value=2.0, min=0.1, max=10.0, step=0.1, description="pois λ")
)
```
